## Scientific background
El Nino Southern Oscillation (ENSO) is a global scale climate phenomenon consisting of two phases: Positive associated with El Nino and negative associated with La Nina. Southeast Asia and Australia associate the El Nino phase with droughts and heatwaves; on the other hand, La Nina corresponds to extreme precipitation (Villafuerte et al, 2015). Whereas, India situated in the South of the Asian continent has not shown a clear affinity with this trend. The strongest recorded El Nino year resulted in excess rainfall, but they were not completely devoid of drought and extreme temperatures during the El Nino years. Out of the 23 El Nino events only 13 were drought years. These climate variations have significant implications for food security.
In the scientific literature, climate change and climate phenomena have been found to impact food security. Cooper et al. (2019) demonstrated that higher precipitation is often correlated with more hunger and stunting in child development. Kang et al. (2009) suggested the contrary, that increased precipitation can lead to higher crop yield. Kang and colleagues also reviewed that crop production can increase with expanding farming areas or intensifying irrigation, but at the cost of environmental degradation. Ringler et al. (2010) projects a general correlation between climate change (especially higher temperatures) and lower calorie availability, childhood malnutrition, and higher food prices.

## Proposal
1. **Background**: India experiences significant impacts from El Nino and La Nina, two climate phenomena that affect precipitation. El Nino is associated with droughts and heatwaves, while La Nina corresponds to extreme precipitation. India's climate has been linked to food security, with increased precipitation often causing hunger and stunting in child development. However, climate change can also lead to increased crop yield and environmental degradation, as well as lower calorie availability, childhood malnutrition, and higher food prices.

2. **Aims**: We want to draw connections between our understanding of ENSO and of food security in a particular region. We will do so by analysing the trend of global mean temperature, analyzing the trend of SST anomalies in the region of Nino 3.4., and examining the relationship between temperature/rainfall and food security

3. **Hypothesis**: The increase in temperature can negatively impact crop quality and excessive rainfall can detrimentally affect crop growth, we anticipate that an increase of global average temperature and rainfall would negatively correlate with food security in India

4. **We will be**:
- running spatio-temporal analysis, to check the general trends and distribution of the differents variables over time;
- exploring standard deviation, correlation, heat map, regression analysis, etc.
- calculating the climatology of monthly averaged data from 1960 to 1990 (period with no trend) of the variables: global 2m temperature, SSTs in Nino 3.4 region, and total precipitation (in India);
- calculating the anomalies of these variables from the data spanning between 1991 to present
- evaluating using statistical methods the relationship between these variables and food security variables in this time period.

5. **Potential societal impact**: A better understanding of the impact of temperature and precipitation in the wet season in India allows better preparation against potential years of increased food insecurity. We would be able to predict the food or aid a region would need when higher temperature or precipitation is forecasted for the year.

```python
# find the monthly climatology and anomaly of total precipitation for India
# Set periods
clim_period = ["1960-01-01T00:00:00.000000000", "1990-12-31T00:00:00.000000000"]
series_period = ["1991-01-01T00:00:00.000000000", "2023-12-31T00:00:00.000000000"]

ds_tp_india_clim = ds_tp_india.sel(time=slice(clim_period[0], clim_period[1])).groupby("time.month").mean(dim="time")

ds_tp_india_mon = ds_tp_india.sel(time=slice(series_period[0], series_period[1])).groupby("time.month")

ds_tp_india_anomaly = ds_tp_india_mon - ds_tp_india_clim

tp_india_plot = ds_tp_india_anomaly.mean(dim=["latitude", "longitude"])
```
